# PROBLEM STATEMENT - 
## We will construct a linear model that explains the relationship that a car's mileage(mpg) has with its other attributes

```python
### car name is not required
df2.drop('car name',axis=1, inplace=True)
```

```python
df2.head()
```

```python
df2['origin'] = df2['origin'].replace({1:'America', 2:"Europe", 3:"Asia"})
```

```python
df2.head()
```

```python
### One-hot Encoding
df2 = pd.get_dummies(df2, columns=['origin'])
```

```python
df2.sample(10)
```

```python
df2.isnull().sum()
```

```python
df2.dtypes
```

### Bivariate Plots

```python
sns.pairplot(df2, diag_kind="kde")
```

### Split Data

```python
### Lets build our linear model
### independent variables
X = df2.drop(['mpg'], axis=1)
### dependent variable
y = df2[['mpg']]
```

```python
### Split X and y into training and test set in 70:30 ratio
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

### Fit Linear Model

```python
model_1 = LinearRegression()
model_1.fit(X_train, y_train)
```

```python
model_1.score(X_train, y_train) # R^2 value for in-sample
```

```python
model_1.score(X_test, y_test) # R^2 value for out of sample
```

## Fit Polynomial Model

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model

poly = PolynomialFeatures(degree=2, interaction_only=True) # good model since it is performing equally good with both train and test data
# poly = PolynomialFeatures(degree=3, interaction_only=True) # overfitting model since it is performing well with train data but bad with test data
X_train2 = poly.fit_transform(X_train)
X_test2 = poly.fit_transform(X_test)

poly_clf = linear_model.LinearRegression()

poly_clf.fit(X_train2, y_train)

#y_pred = poly_clf.predict(X_test2)

#print(y_pred)

#In sample (training) R^2 will always improve with the number of variables
print(poly_clf.score(X_train2, y_train))
```

```python
#Out off sample (testing) R^2 is our measure of success and does improve
print(poly_clf.score(X_test2, y_test))
```

```python
# A model must have the train score and test score difference not greater 
# than 5% to be considered as a correct fit model.
```
