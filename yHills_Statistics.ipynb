# Cardio Good Fitness Case Study - Descriptive Statistics
### The market research team at AdRight is assigned the task to identify the profile of the typical customer for each treadmill product offered by CardioGood Fitness. The market research team decides to investigate whether there are differences across the product lines with respect to customer characteristics. The team decides to collect data on individuals who purchased a treadmill at a CardioGoodFitness retail store during the prior three months. The data are stored in the CardioGoodFitness.csv file.

## The team identifies the following customer variables to study:
### * product purchased, TM195, TM498, or TM798;
### * gender;
### * age, in years;
### * education, in years;
### * relationship status, single or partnered;
### * annual household income ;
### * average number of times the customer plans to use the treadmill each week;
### * average number of miles the customer expects to walk/run each week;
### * and self-rated fitness on an 1-to-5 scale, where 1 is poor shape and 5 is excellent shape.

## Perform descriptive analytics to create a customer profile for each CardioGood Fitness treadmill product line.

```python
# Load the necessary packages

import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
mydata = pd.read_csv("CardioGoodFitness.csv")
```

```python
mydata.head()
```

```python
mydata.shape
```

```python
mydata.dtypes
```

```python
mydata.columns
```

```python
mydata.info()
```

```python
mydata.isnull().sum()
```

```python
mydata.describe() # by default, outputs only for integer datatype(numerical data)
```

```python
mydata.describe(include="all")
```

## Count plot is used to study a categorical feature and gives visual representation

```python
sns.countplot(x="Product",data=mydata)
```

```python
sns.countplot(x="Gender",data=mydata)
```

```python
sns.countplot(x="MaritalStatus",data=mydata)
```

```python
sns.countplot(x="Product",hue="Gender",data=mydata)
```

```python
sns.countplot(x="Product",hue="MaritalStatus",data=mydata)
```

```python
sns.countplot(x="Gender",hue="MaritalStatus",data=mydata)
```

```python
sns.countplot(x="MaritalStatus",hue="Gender",data=mydata)
```

## Histogram can be used to find the distribution of the data

```python
mydata.hist(figsize=(10,20))
plt.show()
```

## Box plot can be used for one categorical and one numerical attributes

```python
sns.boxplot(x="Product", y="Age", data=mydata)
```

```python
# outlier is the value which is less than Q1-1.5*IQR or greater than Q3+1.5*IQR

## From the above boxplot, we can understand that all the outliers are > Q3+1.5*IQR for 3 plots 
```

## Crosstab is used to study categorical features and gives the numerical representation

```python
pd.crosstab(mydata['Product'],mydata['Gender'])
```

```python
pd.crosstab(mydata['Product'],mydata['MaritalStatus'])
```

```python
pd.crosstab(mydata['Gender'],mydata['MaritalStatus'])
```

## Pairplot is used to find the correlation btwn all the attributes

```python
sns.pairplot(mydata, diag_kind='kde')
plt.show()
```

```python
corr = mydata.corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap="YlGnBu")
plt.show()
```

```python
sns.heatmap(corr,cmap="YlGnBu")
plt.show()
```

```python
sns.heatmap(corr,annot=True)
plt.show()
```

```python
sns.heatmap(corr)
plt.show()
```
